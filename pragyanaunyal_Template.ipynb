#### Chart - 1   Distribution of PM2.5 (Histogram)

````python
# Chart - 1 visualization code
plt.figure(figsize=(8,5))

sns.histplot(data=df,
             x='PM2.5',
             bins=30,
             kde=True)

plt.title("Distribution of PM2.5")
plt.xlabel("PM2.5 (µg/m³)")
plt.ylabel("Frequency")

plt.show()
````

##### 1. Why did you pick the specific chart?

A histogram was chosen because PM2.5 is a continuous numerical variable, and histograms are the most appropriate visualization for understanding the distribution of continuous data. Unlike bar charts, which are designed for categorical comparisons, or pie charts, which show proportions, a histogram displays how frequently different PM2.5 concentration ranges occur. It also helps identify the shape of the distribution, such as whether it is symmetric, positively or negatively skewed, and whether there are multiple peaks. The addition of the Kernel Density Estimation (KDE) curve provides a smooth representation of the distribution, making it easier to interpret overall pollution patterns. This makes the histogram the best choice for analyzing the overall behavior of PM2.5 concentrations.

#### Chart - 2  Box Plot of PM10

````python
# Chart - 2 visualization code
plt.figure(figsize=(8,5))

sns.boxplot(x=df["PM10"])

plt.title("Box Plot of PM10")
plt.xlabel("PM10 (µg/m³)")

plt.show()
````

##### 1. Why did you pick the specific chart?

A box plot was selected because its primary strength is identifying outliers while summarizing the distribution of numerical data. Compared to a histogram, which focuses on frequency distribution, a box plot clearly presents the median, quartiles, interquartile range (IQR), and extreme values in a compact form. PM10 measurements may contain unusually high pollution events that are important for environmental analysis, and these can be identified immediately using a box plot. Since one of the key objectives of exploratory data analysis is detecting abnormal observations, a box plot is more suitable than other visualization techniques for this purpose.

This makes it the most appropriate choice for understanding the climatic characteristics of the dataset before examining their influence on pollution.

#### Chart - 4  Number of Records by Country Count Plot


````python
# Chart - 4 visualization code
plt.figure(figsize=(12,6))

sns.countplot(data=df,
              x="Country",
              order=df["Country"].value_counts().index)

plt.title("Number of Air Quality Records by Country")
plt.xlabel("Country")
plt.ylabel("Count")

plt.xticks(rotation=45)

plt.show()
````

##### 1. Why did you pick the specific chart?

A count plot was chosen because Country is a categorical variable, and the objective is to compare the number of observations belonging to each category. Unlike histograms, which are designed for numerical data, or pie charts, which become difficult to interpret when there are many categories, a count plot provides a clear comparison of observation frequencies across countries. It allows easy identification of countries with the highest and lowest number of recorded measurements, making it the most effective visualization for categorical frequency analysis.
