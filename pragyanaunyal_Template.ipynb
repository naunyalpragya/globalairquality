````python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

````

````python
# Load Dataset
df = pd.read_csv("/content/global_air_quality_data_10000.csv")
````

#### Duplicate Values

````python
# Dataset Duplicate Value Count
duplicate_count = df.duplicated().sum()

print(f"Number of Duplicate Records : {duplicate_count}")
````

#### Missing Values/Null Values

````python
# Missing Values/Null Values Count
missing_values = df.isnull().sum()

print(missing_values)
````

````python
# Visualizing the missing values
missing = df.isnull().sum()

plt.figure(figsize=(10,5))

missing.plot(kind='bar')

plt.title("Missing Values in Each Column")
plt.xlabel("Columns")
plt.ylabel("Missing Values")

plt.xticks(rotation=45)

plt.show()
````

### What did you know about your dataset?

The dataset contains 10,000 air quality records collected from major cities across the world in 2023. It includes information on pollutant concentrations (PM2.5, PM10, NO₂, SO₂, CO, and O₃), along with meteorological variables such as temperature, humidity, and wind speed. Data quality checks revealed that the dataset contains no missing (null) values, indicating that it is complete and suitable for exploratory data analysis (EDA) and visualization without requiring imputation. This improves the reliability of the analysis and reduces the need for extensive data cleaning.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the original dataset
df_clean = df.copy()

# Convert the Date column to datetime format
df_clean['Date'] = pd.to_datetime(df_clean['Date'])

# Check for duplicate records
print("Duplicate Records :", df_clean.duplicated().sum())

# Remove duplicate records (if any)
df_clean.drop_duplicates(inplace=True)

# Check for missing values
print("\nMissing Values:")
print(df_clean.isnull().sum())

# Verify data types
print("\nData Types:")
print(df_clean.dtypes)

# Reset the index after cleaning
df_clean.reset_index(drop=True, inplace=True)

# Display the first five rows of the cleaned dataset
df_clean.head()
````
