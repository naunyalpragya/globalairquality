#### Chart - 1   Distribution of PM2.5 (Histogram)

````python
# Chart - 1 visualization code
plt.figure(figsize=(8,5))

sns.histplot(data=df,
             x='PM2.5',
             bins=30,
             kde=True)

plt.title("Distribution of PM2.5")
plt.xlabel("PM2.5 (µg/m³)")
plt.ylabel("Frequency")

plt.show()
````

##### 1. Why did you pick the specific chart?

A histogram was chosen because PM2.5 is a continuous numerical variable, and histograms are the most appropriate visualization for understanding the distribution of continuous data. Unlike bar charts, which are designed for categorical comparisons, or pie charts, which show proportions, a histogram displays how frequently different PM2.5 concentration ranges occur. It also helps identify the shape of the distribution, such as whether it is symmetric, positively or negatively skewed, and whether there are multiple peaks. The addition of the Kernel Density Estimation (KDE) curve provides a smooth representation of the distribution, making it easier to interpret overall pollution patterns. This makes the histogram the best choice for analyzing the overall behavior of PM2.5 concentrations.

##### 2. What is/are the insight(s) found from the chart?

Shows the concentration range where most PM2.5 observations lie.

Detects whether pollution levels are normally distributed or skewed.

Highlights unusually high pollution events.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

Yes. PM2.5 is one of the most harmful air pollutants affecting public health. Understanding its distribution helps governments identify pollution severity, prioritize mitigation strategies, improve healthcare planning, and formulate air-quality policies. Extremely high PM2.5 concentrations may increase healthcare expenditure and reduce workforce productivity.

#### Chart - 6 Average PM2.5 Concentration by Country (Bar Plot)

````python
# Chart - 6 visualization code
plt.figure(figsize=(10,6))

avg_pm25 = df.groupby("Country")["PM2.5"].mean().sort_values(ascending=False)

sns.barplot(x=avg_pm25.index,
            y=avg_pm25.values)

plt.title("Average PM2.5 Concentration by Country")
plt.xlabel("Country")
plt.ylabel("Average PM2.5 (µg/m³)")
plt.xticks(rotation=45)

plt.show()
````

#### Chart - 8 Monthly Average PM2.5 Trend (Line Plot)

````python
# Convert the Date column into datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Check if there are any invalid dates
print("Missing dates after conversion:", df['Date'].isnull().sum())

# Create a new Month column
df['Month'] = df['Date'].dt.month_name()

# Define month order
month_order = ['January', 'February', 'March', 'April', 'May', 'June',
               'July', 'August', 'September', 'October', 'November', 'December']

# Calculate average PM2.5 for each month
monthly_pm25 = df.groupby('Month')['PM2.5'].mean().reindex(month_order)

# Create the figure
plt.figure(figsize=(10,6))

# Plot the line chart
sns.lineplot(
    x=monthly_pm25.index,
    y=monthly_pm25.values,
    marker='o',
    linewidth=2
)

# Add chart title and labels
plt.title('Monthly Average PM2.5 Concentration', fontsize=14)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Average PM2.5 (µg/m³)', fontsize=12)

# Rotate month labels
plt.xticks(rotation=45)

# Add grid
plt.grid(True, linestyle='--', alpha=0.5)

# Display the chart
plt.show()
````

##### 1. Why did you pick the specific chart?

A line plot is the most suitable visualization for showing changes over an ordered variable such as time. Since the objective is to understand how average PM2.5 levels vary across different months, a line chart clearly displays increasing and decreasing pollution trends. Unlike a bar chart, which focuses on comparing categories independently, a line chart effectively highlights continuity and seasonal variation.

##### 2. What is/are the insight(s) found from the chart?

The chart illustrates how average PM2.5 concentrations change throughout the year.

It helps identify months with relatively higher or lower pollution levels.

Seasonal fluctuations suggest that weather conditions and human activities influence air quality.

Recognizing these trends helps in understanding when pollution-control measures may be most necessary.

Yes. Seasonal pollution trends enable governments and environmental agencies to implement preventive measures before high-pollution periods. Healthcare providers can also prepare for increased cases of pollution-related illnesses during months with poor air quality.

#### Chart - 11  Bubble Chart – Effect of Temperature and Humidity on PM2.5

````python
# Chart - 11 visualization code
plt.figure(figsize=(10,6))

plt.scatter(
    x=df['Temperature'],
    y=df['PM2.5'],
    s=df['Humidity']*3,      # Bubble size based on Humidity
    alpha=0.6,
    edgecolors='black'
)

plt.title('Temperature vs PM2.5 (Bubble Size Represents Humidity)')
plt.xlabel('Temperature (°C)')
plt.ylabel('PM2.5 (µg/m³)')

plt.grid(True)

plt.show()
````

##### 1. Why did you pick the specific chart?

A bubble chart was chosen because it allows the visualization of three numerical variables simultaneously in a single graph. While a normal scatter plot only shows the relationship between two variables, a bubble chart adds a third dimension through the size of each bubble. In this visualization, the x-axis represents Temperature, the y-axis represents PM2.5 concentration, and the bubble size represents Humidity. This makes it easier to understand how weather conditions collectively influence air pollution rather than analyzing each variable separately. It provides a richer and more comprehensive view of the dataset than traditional two-variable charts.

##### 2. What is/are the insight(s) found from the chart?

The chart shows how PM2.5 concentrations vary across different temperature ranges.

Bubble sizes reveal whether higher humidity is associated with higher or lower PM2.5 concentrations.

It helps identify clusters where temperature, humidity, and pollution occur together.

Outliers indicate unusual weather conditions or pollution events that may require further investigation.
